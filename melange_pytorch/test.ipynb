```python
import numpy as np
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
loss = np.loadtxt('test_loss.txt')
plt.plot(loss)
plt.title('Test Loss')
plt.xlabel('Epochs')

plt.subplot(1, 2, 2)
Y_psi = np.loadtxt('Y_psi.txt')
output_psi = np.loadtxt('output_psi.txt')

plt.plot(Y_psi, output_psi, 'o', markersize=1)
plt.xlabel('Y_psi')
plt.ylabel('output_psi')

#calculate correlation coefficient
corr = np.corrcoef(Y_psi, output_psi)[0, 1]
plt.title('Correlation coefficient: ' + str(corr))
```
