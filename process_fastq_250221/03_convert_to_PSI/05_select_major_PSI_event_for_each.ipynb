# Select major alternative ss events for each: WT.

```r
library(tidyverse)
library(vroom)
library(data.table)
library(pheatmap)
library(preprocessCore)
library(purrr)
library(RColorBrewer)

WT_count <- fread("/mnt/dawnccle2/processed_data/reprocess_250221/count_normalized_v4_merged/WT_all_samples_raw_counts.csv")

WT_count <- WT_count %>% 
    filter(mode != "UNSPLICED") %>%
    group_by(sample, index) %>% 
    mutate(total_count = sum(count_scaled)) %>% 
    mutate(count_perc = count_scaled / total_count) %>% 
    ungroup() 

# Get the skipped count for each sample. 
skipped_count <- WT_count %>% 
    filter(mode == "SKIPPED") %>% 
    group_by(sample, index) %>% 
    summarise(skipped_count = sum(count_scaled),
    skipped_perc = skipped_count / total_count) %>% 
    ungroup() 

# Merge the skipped count with the major event PSI. 
WT_count <- WT_count %>% 
    left_join(skipped_count, by = c("sample", "index")) %>% 
    # Set NA to 0. 
    mutate(skipped_count = ifelse(is.na(skipped_count), 0, skipped_count)) %>% 
    mutate(skipped_perc = ifelse(is.na(skipped_perc), 0, skipped_perc)) 
```

```r
WT_included <- WT_count %>% 
    filter(total_count >= 30) %>%
    filter(mode == "INCLUDED") 

# Get the major included event for each index. 
major_included_event <- WT_included %>% 
    group_by(index, offset) %>% 
    summarise(average_count_perc = mean(count_perc)) %>% 
    ungroup() %>% 
    group_by(index) %>% 
    arrange(desc(average_count_perc)) %>% 
    slice(1) %>% 
    ungroup() 

# Write this to a csv file.
write_csv(major_included_event, "/mnt/dawnccle2/processed_data/reprocess_250221/count_normalized_v4_merged/WT_included_major_event_by_index.csv")
```

```r
# Filter the list of events.
major_included_event_filtered <- major_included_event %>% arrange((average_count_perc)) %>% 
    filter(average_count_perc > 0.01) %>% 
    separate(offset, into = c("upstream_offset", "downstream_offset", "const_offset"), sep = ":", remove = FALSE) %>% 
    mutate(upstream_offset = as.numeric(upstream_offset),
    downstream_offset = as.numeric(downstream_offset),
    const_offset = as.numeric(const_offset)) %>% 
    filter(upstream_offset < 200 & downstream_offset < 200)

write_csv(major_included_event_filtered, "/mnt/dawnccle2/processed_data/reprocess_250221/count_normalized_v4_merged/WT_included_major_event_by_index_filtered.csv")

# For each "major event", get the list of events that are not major. Where the table will have 3 columsn: index, offset_major, offset_minor.
WT_included_index_offset <- WT_included %>% 
    mutate(index_offset = paste(index, offset, sep = "__")) %>% 
    select(index_offset, index, offset) %>% 
    distinct()

WT_major_events <- major_included_event_filtered %>% 
    mutate(index_offset = paste(index, offset, sep = "__")) %>% 
    select(index_offset, index, offset) %>% 
    distinct()

# Merge all_and_major_events. 
WT_included_all_event_pairs <- WT_major_events %>% 
    left_join(WT_included_index_offset, by = "index") %>%
    filter(index_offset.x != index_offset.y) %>% 
    rename(index_offset_major = index_offset.x, index_offset_minor = index_offset.y) %>% 
    rename(offset_major = offset.x, offset_minor = offset.y)

write_csv(WT_included_all_event_pairs, "/mnt/dawnccle2/processed_data/reprocess_250221/count_normalized_v4_merged/WT_included_all_unique_event_pairs.csv")
```
